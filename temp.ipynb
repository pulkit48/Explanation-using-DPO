```python
import os

# Base sbt command template
base_command = (
    'sbt "runMain org.allenai.pdffigures2.FigureExtractorBatchCli '
    '{input_path} --dpi 300 -s {stats_path} -m {figures_path} -d {data_path}"'
)

# Root folder for the dataset
folder_path = "C:/Users/pulki/Downloads/Fujitsu_Work/dataset/test-C/SPIQA_testC"

# Iterate through subdirectories in the folder path
directories = [d for d in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, d))]
for dir_name in directories:
    folder = os.path.join(folder_path, dir_name)
    print(f"Processing folder: {folder}")

    # Check for subdirectory that matches the folder name (1603.03833v4/1603.03833v4)
    matching_subdirs = [item for item in os.listdir(folder) if item.startswith(dir_name)]
    for subdir in matching_subdirs:
        folder2 = os.path.join(folder, subdir)
        print(f"Found matching subdirectory: {folder2}")

        # Define paths for stats, figures, and data
        stats_path = os.path.join(folder, "data", "stats.json")
        figures_path = os.path.join(folder, "figures/")
        data_path = os.path.join(folder, "data/")

        # Format the command
        command = base_command.format(
            input_path=folder2,
            stats_path=stats_path.replace("\\", "/"),  # Ensure paths are forward slashes for sbt
            figures_path=figures_path.replace("\\", "/"),
            data_path=data_path.replace("\\", "/"),
        )

        # Run the command
        print(f"Running command:\n{command}")
        os.system(command)

    # break
```

```python
import os
import shutil

folder_path="C:\\Users\\pulki\\Downloads\\Fujitsu_Work\\dataset\\test-B\\SPIQA_testB"
directories = [d for d in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, d))]
for dir in directories:
    folder=folder_path+"\\"+dir
    print(folder)
    os.mkdir(folder+"\\data")
    os.mkdir(folder+"\\figures")
    for item in os.listdir(folder):
        if item.startswith(dir):
            folder2 = os.path.join(folder, item)
            file_to_keep_path=item+".pdf"
            for item in os.listdir(folder2):
                item_path = os.path.join(folder2, item)
            
                # Skip the PDF file to keep
                if item == file_to_keep_path:
                    continue

                # Remove directories
                if os.path.isdir(item_path):
                    shutil.rmdir(item_path)  # Use shutil.rmtree for non-empty directories
                # Remove files
                elif os.path.isfile(item_path):
                    os.remove(item_path)
        
    
    
            
    
            
            
```
