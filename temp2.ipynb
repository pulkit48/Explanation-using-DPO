```python
# Extracting lables from the tex file using captions from the json file of figures and tables
import json
import re
from rapidfuzz import fuzz
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer
# Paths to the files
def updation(json_data,tex_file):
    # json_data = r'C:\Users\pulki\Downloads\Fujitsu_Work\dataset\test-A\SPIQA_testA\1603.00286v5\data\1603.00286v5.json'
    # tex_file = r'C:\Users\pulki\Downloads\Fujitsu_Work\dataset\test-A\SPIQA_testA\1603.00286v5\tex_files\redivision-18.tex'
    # Load the JSON data (list of dictionaries)
    with open(json_data, 'r',encoding="iso-8859-1") as json_file:
        json_content = json.load(json_file)
    # Open and read the .tex file
    with open(tex_file, 'r',encoding="iso-8859-1") as tex_file:
        tex_content = tex_file.read()
    # Function to normalize LaTeX text (removing \textit, special characters, and commands)
    def normalize_latex(text):
        # Remove LaTeX math mode delimiters ($...$), but retain the content between them
        text = re.sub(r'\$', '', text)  # Removes $ symbols but keeps the enclosed content
        # Normalize quotation marks
        text = text.replace("``", '"').replace("''", '"')  # Convert LaTeX-style quotes to standard quotes
        text = text.replace("“", '"').replace("”", '"')    # Convert curly quotes to standard quotes
        # Remove LaTeX formatting commands, preserving text within braces
        text = re.sub(r'\\[a-zA-Z]+{([^{}]*)}', r'\1', text)  # Handles \textit{...}, \textbf{...}, etc.
        text = re.sub(r'\\[a-zA-Z]+', '', text)  # Remove standalone LaTeX commands (e.g., \alpha)
        text = re.sub(r'\s+', ' ', text)  # Normalize whitespace
        return text.strip()
    # Iterate through each dictionary in the JSON list
    for item in json_content:
        caption_from_json = item.get('caption', '')  # Get the caption
        # Remove any prefix (like "Figure 10:") from the caption
        # stripped_caption = re.sub(r'^.*?:\s*', '', caption_from_json)  # Strip text before colon
        normalized_caption = normalize_latex(caption_from_json)
        # Updated regex to handle deeply nested braces in \caption{}
        caption_pattern = re.compile(r'\\caption\{((?:[^{}]|{(?:[^{}]|{[^{}]*})*})*)\}.*?\\label\{(.*?)\}', re.DOTALL)
        matches = caption_pattern.findall(tex_content)
            
        # Initialize variables to track best match
        best_similarity = 0
        best_label = None
        # Compare with all matches to find the highest similarity
        for raw_caption, label in matches:
            # Normalize the caption inside the \caption{} block
            normalized_tex_caption = normalize_latex(raw_caption)
            # Calculate similarity using fuzzy matching
            similarity = fuzz.ratio(normalized_caption, normalized_tex_caption)
            # similarity = fuzz.partial_ratio(normalized_caption, normalized_tex_caption)
            # Update best match if current similarity is higher
            if similarity > best_similarity:
                best_similarity = similarity
                best_label = label
        # Only assign the label if we found a match with similarity above threshold
        if best_similarity >= 60:
            item['label'] = best_label
            # print(f"Caption: {caption_from_json}")
            # print(f"normalized_tex_caption: {normalized_tex_caption}")
            # print(f"Normalized Caption: {normalized_caption}")
            # print(f"Found label: {label}")
            # print('-' * 50)
        else:
            item['label'] = None
            # print(f"No matching caption or label found for caption: {caption_from_json}")
            # print('-' * 50)
            
    new_json_data_path = json_data.replace('.json', '_updated.json')
    # Save the updated JSON data to a new file
    with open(new_json_data_path, 'w') as new_json_file:
        json.dump(json_content, new_json_file, indent=4)
        
    print(f"Updated JSON data saved to {new_json_data_path}")
    
    # has context menu
```

```python
# optimised code for pdf processing to get the bounding box of the text
import re
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer
from rapidfuzz import fuzz
from collections import Counter

class PDFProcessor:
    def __init__(self, pdf_path):
        """
        Initialize the PDF processor by extracting all text elements once.
        
        Args:
            pdf_path (str): Path to the PDF file
        """
        self.text_elements = []
        self._extract_all_text_elements(pdf_path)
    
    def _extract_all_text_elements(self, pdf_path):
        """
        Extract all text elements and their bounding boxes from the PDF.
        
        Args:
            pdf_path (str): Path to the PDF file
        """
        for page_layout in extract_pages(pdf_path):
            for element in page_layout:
                if isinstance(element, LTTextContainer):
                    self.text_elements.append({
                        'page_id': page_layout.pageid,
                        'text': element.get_text(),
                        'bbox': element.bbox
                    })

    def extract_plain_text(self, original_text):
        """
        Extract only human-readable portions of text.
        
        Args:
            original_text (str): The original LaTeX-formatted text
        
        Returns:
            list: List of plain text fragments
        """
        plain_text_fragments = re.findall(r'(?:^|[^\\$\\{}])([a-zA-Z0-9\s,.\'";!?-]+)(?:$|[^\\$\\{}])', original_text)
        plain_text_fragments = [fragment.strip() for fragment in plain_text_fragments if fragment.strip()]
        return plain_text_fragments

    def find_text_bbox_fuzzy(self, search_text):
        """
        Extract top 5 longest text fragments from the original text and locate their bounding boxes in the PDF.

        Args:
            pdf_path (str): Path to the PDF file.
            original_text (str): The original text containing potential fragments to search.

        Returns:
            str: Description of the most common bounding box if duplicates exist, or the first bounding box found.
        """
        # Normalize the search text (lowercase, remove extra spaces/line breaks)
        def normalize_text(text):
            return " ".join(text.lower().split())

        normalized_search_text = normalize_text(search_text)
        highest_score = 0
        best_page_id = None
        best_bbox = None

        for element in self.text_elements:
            normalized_text = normalize_text(element['text'])

            # Perform the operation only if search text length <= extracted text length
            if len(normalized_search_text) <= len(normalized_text) and len(normalized_search_text.split())>=3:
                # Calculate similarity score
                score = fuzz.partial_ratio(normalized_search_text, normalized_text)

                if score > highest_score and score >=60:
                    highest_score = score
                    best_page_id = element['page_id']
                    best_bbox = element['bbox']
                    
        return best_page_id, best_bbox

    def process_text_fragments(self, original_text):
        """
        Process text fragments and find their bounding boxes.
        
        Args:
            original_text (str): Original text to process
        
        Returns:
            dict: Dictionary containing page_id and bbox
        """
        plain_text_fragments = self.extract_plain_text(original_text)
        sorted_fragments = sorted(plain_text_fragments, key=len, reverse=True)[:5]
        
        bbox_results = []
        for fragment in sorted_fragments:
            page_id, bbox = self.find_text_bbox_fuzzy(fragment)
            if bbox:
                bbox_results.append((page_id, bbox))
        
        if not bbox_results:
            return {"page_id": None, "bbox": None}
        
        bbox_counter = Counter(bbox for _, bbox in bbox_results)
        most_common_bbox = bbox_counter.most_common(1)
        
        if most_common_bbox and most_common_bbox[0][1] > 1:
            return {"page_id": bbox_results[0][0], "bbox": bbox_results[0][1]}
        else:
            return {"page_id": bbox_results[0][0], "bbox": bbox_results[0][1]}

# Example usage
# original_text = '''Figure~\\ref{fig:casestudy4} shows the effectiveness of leveraging image-patch embeddings to enhance the comprehension of entity representation in cross-page scenarios. RoI-based models only learn the relations between document entity embeddings without considering the contextual information of input document images.'''

# pdf_path = "/mnt/commonfolder/genAIllm/prasad/Data_sources/MMVQA/2404.12720v1/2404.12720v1.pdf"
# pdf_processor = PDFProcessor(pdf_path)
# result = pdf_processor.process_text_fragments(original_text)
# print(result)
```

```python
import re
from pathlib import Path
import json
from rapidfuzz import fuzz

def is_commented_line(line):
    """
    Check if a line is a LaTeX comment line.
    Handles both full-line comments and inline comments.
    
    Args:
        line (str): Line to check
        
    Returns:
        bool: True if the line is a comment or part of a comment block
    """
    # Remove leading whitespace
    stripped_line = line.lstrip()
    
    # Check if line starts with % (full line comment)
    if stripped_line.startswith('%'):
        return True
        
    # Check if this line is part of a comment block
    if stripped_line.startswith('\\begin{comment}') or stripped_line.startswith('\\end{comment}'):
        return True
    
    return False

def extract_figure_text(line):
    """
    Extract only the non-commented portion of a line if it contains an inline comment.
    
    Args:
        line (str): Line to process
        
    Returns:
        str: Line with comments removed
    """
    # Split the line at % but not at \% (escaped percentage)
    parts = re.split(r'(?<!\\)%', line)
    return parts[0].rstrip()

def extract_references(tex_file_path):
    """
    Extract non-commented lines that contain references to figures or tables from a LaTeX file.
    Specifically designed for academic conference papers (ACL, ICML, IJCAI, etc.)
    
    Args:
        tex_file_path (str): Path to the .tex file
        
    Returns:
        list: List of tuples containing (reference type, line text, line number)
    """
    # Read the tex file
    with open(tex_file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    
    # Comprehensive patterns for academic papers
    figure_patterns = [
        r'\\ref\{[Ff]ig:.*?\}',          # \ref{fig:...} or \ref{Fig:...}
        r'\\ref\{[Ff]igure:.*?\}',       # \ref{figure:...} or \ref{Figure:...}
        r'\\ref\{[Ff]ig\.*?\}',          # \ref{fig.}
        r'[Ff]igure\s*\\ref\{.*?\}',     # Figure \ref{...}
        r'[Ff]ig\.\s*\\ref\{.*?\}',      # Fig. \ref{...}
        r'[Ff]igs?\.\s*\\ref\{.*?\}',    # Figs. \ref{...}
        r'[Ff]igures?\s*\\ref\{.?\}(\s,\s*\\ref\{.?\})',  # Figure \ref{...}, \ref{...}
        r'[Ff]igs?\.\s*\\ref\{.?\}(\s,\s*\\ref\{.?\})',   # Figs. \ref{...}, \ref{...}
        r'[Ff]igs?\.\s*\\ref\{.?\}\s(-|--|\s+to\s+)\\ref\{.*?\}',  # Figs. \ref{...}--\ref{...}
        r'\\figurename\s*\\ref\{.*?\}',   # \figurename \ref{...}
        r'\\figref\{.*?\}',               # \figref{...}
        r'\\ref\{[Ff]ig:.*?:[a-zA-Z]\}',  # \ref{fig:name:a}
        r'\\subref\{.*?\}'                # \subref{...}
    ]
    
    table_patterns = [
        r'\\ref\{[Tt]ab:.*?\}',          # \ref{tab:...} or \ref{Tab:...}
        r'\\ref\{[Tt]able:.*?\}',        # \ref{table:...} or \ref{Table:...}
        r'\\ref\{[Tt]ab\.*?\}',          # \ref{tab.}
        r'[Tt]able\s*\\ref\{.*?\}',      # Table \ref{...}
        r'[Tt]ab\.\s*\\ref\{.*?\}',      # Tab. \ref{...}
        r'[Tt]abs?\.\s*\\ref\{.*?\}',    # Tabs. \ref{...}
        r'[Tt]ables?\s*\\ref\{.?\}(\s,\s*\\ref\{.?\})',  # Table \ref{...}, \ref{...}
        r'[Tt]abs?\.\s*\\ref\{.?\}(\s,\s*\\ref\{.?\})',  # Tabs. \ref{...}, \ref{...}
        r'[Tt]abs?\.\s*\\ref\{.?\}\s(-|--|\s+to\s+)\\ref\{.*?\}',  # Tabs. \ref{...}--\ref{...}
        r'\\tablename\s*\\ref\{.*?\}',    # \tablename \ref{...}
        r'\\tabref\{.*?\}'                # \tabref{...}
    ]
    
    algorithm_patterns = [
        r'\\ref\{[Aa]lg:.*?\}',           # \ref{alg:...}
        r'[Aa]lgorithm\s*\\ref\{.*?\}',   # Algorithm \ref{...}
        r'[Aa]lg\.\s*\\ref\{.*?\}'        # Alg. \ref{...}
    ]
    
    results = []
    in_comment_block = False
    
    # Process each line
    for line_num, line in enumerate(lines, 1):
        # Check if we're entering or leaving a comment block
        if '\\begin{comment}' in line:
            in_comment_block = True
            continue
        elif '\\end{comment}' in line:
            in_comment_block = False
            continue
            
        # Skip if we're in a comment block or if the line is commented
        if in_comment_block or is_commented_line(line):
            continue
            
        # Remove inline comments if present
        line = extract_figure_text(line)
        
        if not line.strip():  # Skip empty lines
            continue
            
        # Check for figure references
        for pattern in figure_patterns:
            if re.search(pattern, line):
                results.append(('Figure', line.strip(), line_num))
                break
                
        # Check for table references
        for pattern in table_patterns:
            if re.search(pattern, line):
                results.append(('Table', line.strip(), line_num))
                break
                
        # # Check for algorithm references
        # for pattern in algorithm_patterns:
        #     if re.search(pattern, line):
        #         results.append(('Algorithm', line.strip(), line_num))
        #         break
    
    return results

def main(pdf_file_path,tex_file_path,json_file_path,output_json_path)  :
    """
    Main function to process a tex file and append referenced text to JSON data.
    """
    # tex_file_path=r"C:/Users/pulki/Downloads/Fujitsu_Work/dataset/test-A/SPIQA_testA/1603.00286v5/tex_files/redivision-18.tex"
    # json_file_path=r"C:/Users/pulki/Downloads/Fujitsu_Work/dataset/test-A/SPIQA_testA/1603.00286v5/data/1603.00286v5_updated.json"
    # pdf_file_path=r"C:/Users/pulki/Downloads/Fujitsu_Work/dataset/test-A/SPIQA_testA/1603.00286v5/1603.00286v5/1603.00286v5.pdf"
    # output_json_path=r"C:/Users/pulki/Downloads/Fujitsu_Work/dataset/test-A/SPIQA_testA/1603.00286v5/data/1603.00286v5_final.json"
    # pdf_file_path = '/mnt/commonfolder/genAIllm/prasad/Data_sources/MMVQA/2404.12720v1/2404.12720v1.pdf'
    # tex_file_path = '/mnt/commonfolder/genAIllm/prasad/Data_sources/MMVQA/tex_files/main_final.tex'
    # json_file_path = '/mnt/commonfolder/genAIllm/prasad/Data_sources/MMVQA/data/2404.12720v1_updated.json'
    # output_json_path = '/mnt/commonfolder/genAIllm/prasad/Data_sources/MMVQA/data/2404.12720v1_final.json'

    try:
        # Load the JSON data
        with open(json_file_path, 'r') as json_file:
            json_data = json.load(json_file)

        # Extract references from the .tex file
        references = extract_references(tex_file_path)

        if not references:
            print("No figure, table, or algorithm references found in the document.")
            return

        # Append references to JSON data
        for item in json_data:
            label = item.get('label')
            if label:
                matching_references = []
                for ref_type, line_text, line_num in references:
                    if f'{{{label}}}' in line_text and ref_type== item.get('figType'): # Check if the label is present in the line and the reference type matches
                        matching_references.append({
                            'reference_type': ref_type,
                            'line_text': line_text,
                            'line_number': line_num
                        })
                        
                pdf_processor = PDFProcessor(pdf_file_path)
                if matching_references:
                    item['reference_text'] = matching_references
                    
                    # Append bounding box data for each reference
                    for ref in matching_references:
                        bbox_result = pdf_processor.process_text_fragments(ref['line_text'])
                        ref['text_page_no'] = bbox_result['page_id']
                        ref['text_bbox'] = bbox_result['bbox']
                    
        # Remove items without 'reference_text'
        json_data = [item for item in json_data if 'reference_text' in item]
        # Remove 'imageText' key if present in JSON items
        for item in json_data:
            if 'imageText' in item:
                del item['imageText']

        # Save the updated JSON data to a new file
        with open(output_json_path, 'w') as output_file:
            json.dump(json_data, output_file, indent=4)

        print(f"Updated JSON data saved to {output_json_path}")

    except FileNotFoundError:
        print(f"Error: File '{tex_file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")


```
